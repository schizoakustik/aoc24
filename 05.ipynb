```python
class UpdateChecker:
    def __init__(self, update: list[str], number: int) -> None:
        self.update = update
        self.clear = True
        self.number = number + 1

    def check(self) -> int:
        for i, page in enumerate(self.update):
            for e in self.update[i:]:
                if e in after_dict[page]:
                    self.clear = False
        if self.clear:
            # print(f"[âœ“][#{self.number}] {self.update}")
            return int(self.update[len(self.update)//2])
        # incorrects.append(self.update)
        # print(f"[!][#{self.number}] is not clear.")
        return False
        
    def fix(self) -> None:
        # print(f"[i] Trying to fix update #{self.number}...")
        for i, page in enumerate(self.update):
            for e in self.update[i:]:
                if e in after_dict[page]:
                    # print(f"[!][#{self.number}] {page} must be after {e} at index {self.update.index(e)}")
                    self.update.insert(self.update.index(e), self.update.pop(i))
            for e in self.update[:i]:
                if e in before_dict[page]:
                    # print(f"[i][#{self.number}] {page} at index {i} must be before {e} at index {self.update.index(e)}.")
                    self.update.insert(self.update.index(e), self.update.pop(i))
```
